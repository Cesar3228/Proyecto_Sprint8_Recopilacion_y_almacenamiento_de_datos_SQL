## Prueba de Hipótesis

LA hipótesis es:
- H0= La duración promedio es la misma tanto en días lluviosos y días sin lluvia.
- H1= La duración promedio cambia en los días lluviosos.

<div class="alert alert-block alert-success"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

Buen trabajo formulando la hipótesis! Es fundamental que sea claro lo que se busca analizar.

</div>

```python
df2 = pd.read_csv('/datasets/project_sql_result_07.csv')
```

```python
df2
```

```python
df2.info()
```

```python
df2.describe(include='all')
```

```python
df2['weather_conditions'].value_counts()
```

```python
good=df2[df2['weather_conditions']=='Good']['duration_seconds']
good
```

```python
bad=df2[df2['weather_conditions']=='Bad']['duration_seconds']
bad
```

```python
alpha=.05
```

```python
from scipy.stats import levene
```

```python
# Prueba de Levene
stat, p= levene(good,bad,center='median')
print(f"Prueba Levene: stat={stat:.3f}, p={p:.4f}")
```

```python
if p < alpha:
    print("Rechazamos H0 → Las varianzas son distintas.")
else:
    print("No rechazamos H0 → No hay evidencia suficiente de que las varianzas sean diferentes.")
```

<div class="alert alert-block alert-success"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

Perfecto! Comprobar las varianzas te asegura un uso correcto de las pruebas estadísticas.

</div>

```python
t_stat, p_val= st.ttest_ind(good,bad,equal_var=True)
print(f"t = {t_stat:.3f}, p = {p_val:.4f}")
```

```python
if p_val < alpha:
    print("Rechazamos H0 → La duración promedio SÍ cambia los sábados lluviosos.")
else:
    print("No rechazamos H0 → No hay evidencia suficiente de que cambie la duración promedio.")
```

La conclusión de la prueba es que la duración promedio de los viajes en días lluviosos si cambia.
